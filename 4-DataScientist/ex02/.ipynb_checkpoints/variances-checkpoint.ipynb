```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Train_knight.csv")
```

```python
# Select numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Delete columns with variance 0
df_numeric = df[numeric_columns]
df_numeric = df_numeric.loc[:, df_numeric.var() != 0]
```

```python
# Standarize data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df_numeric)
```

```python
# Apply PCA
pca = PCA()
pca.fit(data_scaled)

# Calculate explained variance
variance_percentages = pca.explained_variance_ratio_ * 100

# Calculate cumulative variance
cumulative_variances = np.cumsum(variance_percentages)

# how many components are needed to reach 90% variance
components_needed = np.where(cumulative_variances >= 90)[0][0] + 1
```

```python
print("Variances (Percentage):")
print(variance_percentages)
print("\nCumulative Variances (Percentage):")
print(cumulative_variances)
print(f"\nWe need {components_needed} components to reach 90% of the total variance.")
```

```python
# Create plot
plt.figure(figsize=(12, 6))
plt.plot(range(1, len(cumulative_variances) + 1), cumulative_variances, 'bo-')
plt.axhline(y=90, color='r', linestyle='--')
plt.axvline(x=components_needed, color='g', linestyle='--')
plt.xlabel('Nmero de componentes')
plt.ylabel('Varianza acumulativa (%)')
plt.title('Varianza acumulativa por componente')
plt.grid(True)

# Aadimos una anotacin para el punto de 90%
plt.annotate(f'90% en {components_needed} componentes', 
             xy=(components_needed, 90),
             xytext=(components_needed + 1, 85),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.tight_layout()
plt.show()
```
