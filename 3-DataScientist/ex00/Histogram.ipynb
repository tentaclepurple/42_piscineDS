```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
try:
    df = pd.read_csv("../Test_knight.csv")
    print(df.head())
except Exception as e:
    print(e)
    
```

```python
    print(df.describe())
```

```python
plt.figure(figsize=(20, 20))
plt.subplots_adjust(hspace=0.5, wspace=0.3)

# Create a histogram for each feature
for i, column in enumerate(df.columns, 1):
    plt.subplot(6, 5, i)
    plt.hist(df[column], bins=20, edgecolor='none', color='lightgreen')
    plt.xlabel('')
    plt.title(column)
    
    # Add a legend that says "knight"
    plt.legend(['knight'], loc='upper right', fontsize='small', edgecolor='none')

plt.tight_layout()
plt.show()
```

```python
try:
    df = pd.read_csv("../Train_knight.csv")
    print(df.head())
except Exception as e:
    print(e)
```

```python
# make 2 df for jedis and siths
jedi_data = df[df['knight'] == 'Jedi']
sith_data = df[df['knight'] == 'Sith']
```

```python
print("JEDI\n", jedi_data.describe())
print()
print("SITH\n", sith_data.describe())
```

```python

# Configure plot
plt.figure(figsize=(20, 20))
plt.subplots_adjust(hspace=0.5, wspace=0.3)

# Create one histogram for each feature
for i, column in enumerate(df.columns[df.columns != 'knight'], 1):
    plt.subplot(6, 5, i)
    
    # Create overlap histograms
    plt.hist(jedi_data[column], bins=30, alpha=0.5, color='dodgerblue', label='Jedi')
    plt.hist(sith_data[column], bins=30, alpha=0.5, color='red', label='Sith')
    
    plt.title(column)
    plt.xlabel('')
    
    # Add legend
    plt.legend(loc='upper right', fontsize='small')

plt.tight_layout()
plt.show()
```
