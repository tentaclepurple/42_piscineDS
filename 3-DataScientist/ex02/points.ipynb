```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('../Train_knight.csv')
test_df = pd.read_csv('../Test_knight.csv')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 13))

# Gráfico 1: Train data con clusters separados
sns.scatterplot(data=train_df, x='Empowered', y='Stims', hue='knight', palette={"Jedi": "blue", "Sith": "red"}, ax=axs[0, 0])
axs[0, 0].set_title('Train Data: Empowered vs Stims (Clustered)')

# Gráfico 2: Train data con clusters mezclados
sns.scatterplot(data=train_df, x='Survival', y='Deflection', hue='knight', palette={"Jedi": "blue", "Sith": "red"}, ax=axs[0, 1])
axs[0, 1].set_title('Train Data: Survival vs Deflection (Mixed)')

# Gráfico 3: Test data sin diferenciar clusters
sns.scatterplot(data=test_df,x='Empowered', y='Stims', color='green', ax=axs[1, 0], label='Knight')
axs[1, 0].set_title('Test Data: Empowered vs Stims (Undifferentiated)')

# Gráfico 4: Test data sin diferenciar clusters
sns.scatterplot(data=test_df, x='Survival', y='Deflection', color='green', ax=axs[1, 1], label='Knight')
axs[1, 1].set_title('Test Data: Survival vs Deflection (Undifferentiated)')

plt.tight_layout()
plt.show()
```
